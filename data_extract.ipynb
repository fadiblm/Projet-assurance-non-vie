```python
import os
import pyreadr

folder = "/home/onyxia/Projet-assurance-non-vie/data"

for file in os.listdir(folder):
    if file.endswith(".rda") or file.endswith(".RData"):
        filepath = os.path.join(folder, file)
        print(f"\n--- Lecture de {file} ---")

        result = pyreadr.read_r(filepath)

        for name, obj in result.items():
            try:
                output_name = f"{name}.csv"
                output_path = os.path.join(folder, output_name)

                obj.to_csv(output_path, index=False, encoding="utf-8-sig")

                print(f"  ✓ Exporté : {output_name}")
            except Exception as e:
                print(f"  ⚠️ Impossible d'exporter {name} : {e}")
```
